```python
password = input("Enter your password: ")

length = len(password)

has_upper = False
has_lower = False
has_digit = False

for char in password:
    if char.isupper():
        has_upper = True
    if char.islower():
        has_lower = True
    if char.isdigit():
        has_digit = True

if length >= 8 and has_upper and has_lower and has_digit:
    print("Strong Password")
elif length >= 6:
    print("Medium Password")
else:
    print("Weak Password")
```

```python
#accessing dictionary keys, values and items
my_dict = {"a": 1, "b": 2, "c": 3}

# Get all keys
print(my_dict.keys())     # dict_keys(['a', 'b', 'c'])

# Get all values
print(my_dict.values())   # dict_values([1, 2, 3])

# Get all key-value pairs
print(my_dict.items())    # dict_items([('a', 1), ('b', 2), ('c', 3)])

# Check if key exists
print("a" in my_dict)     # True
print("z" in my_dict)     # False

# Get length
print(len(my_dict))       # 3
```

```python
scores = {"Rahul": 87, "Sneha": 95, "Vikram": 64, "Priya": 92}

# 1. Loop over keys (most common)
for name in scores:
    print(name, "→", scores[name])
    # 2. Loop over keys explicitly
for name in scores.keys():
    print(name)
    # 3. Loop over values
for mark in scores.values():
    print(mark, end=" ")
    # 4. Loop over both key and value (very powerful!)
for name, mark in scores.items():
    print(f"{name:8} : {mark}")
    if mark >= 90:
        print("   → Outstanding!")
```

```python
def count_characters(text):
    char_count = {}
    
    for char in text:
        if char != " ":   # Ignore spaces
            if char in char_count:
                char_count[char] += 1
            else:
                char_count[char] = 1
                
    return char_count


# Example
result = count_characters("hello world")
print(result)
```

```python
nested_list = [1, [2, 3], [4, [5, 6]], 7]

def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

print(flatten(nested_list))
```

```python
nested_list = [1, [2, 3], [4, [5, 6]], 7]

flat_list = []

for item in nested_list:
    if isinstance(item, list):
        for sub in item:
            flat_list.append(sub)
    else:
        flat_list.append(item)

print(flat_list)
```

```python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

anagrams = {}

for word in words:
    key = ''.join(sorted(word))
    
    if key in anagrams:
        anagrams[key].append(word)
    else:
        anagrams[key] = [word]

print(list(anagrams.values()))
```

```python
from collections import defaultdict

words = ["eat", "tea", "tan", "ate", "nat", "bat"]

anagrams = defaultdict(list)

for word in words:
    key = ''.join(sorted(word))
    anagrams[key].append(word)

print(list(anagrams.values()))
```
